## Generating data

```python
idx = 11
channels = {'H1':'INJ', 'L1':'INJ', 'V1':'INJ'}
outdir = f'injection_{idx}'
injection_file = f"../../benchmarking/non_precessing_random/injections_XPHM_2000_fiducial_pop/injection_{idx}/injection_parameters.json"

asds = {
    'H1': '/home/mukesh.singh/work/Simple-PE/benchmarking/asds/aligo_O4high.txt',
    'L1': '/home/mukesh.singh/work/Simple-PE/benchmarking/asds/aligo_O4high.txt',
    'V1': '/home/mukesh.singh/work/Simple-PE/benchmarking/asds/avirgo_O4high_NEW.txt'
}
psds = {}

delta_f = 0.0625
f_low = 15.0
seed = 12345
gaussian_noise = False
```

```python
if not os.path.isdir(outdir):
    os.mkdir(outdir)
_strain, _channels = {}, {}
for ifo, value in channels.items():
    if "inj" in value.lower():
        if not os.path.isfile(injection_file):
            raise FileNotFoundError(
                f"Unable to find file: {opts.injection}"
            )
        with open(injection_file, "r") as f:
            injection_params = json.load(f)
        # high frequency content in the psd used for generating 
        # gaussian noise is set by the sampling rate of the data
        if 'delta_t' in injection_params:
            f_high = 1 / 2 / injection_params["delta_t"]
        else:
            raise KeyError(
                "delta_t must be passed to create an injection"
            )
        psd = io.load_psd_from_file(
            psds, asds, delta_f, f_low, f_high
        )
        _strain[ifo], _channels[ifo] = get_injection_data(
            outdir, injection_params, ifo, psds, seed, 
            gaussian_noise
        )
    else:
        raise ValueError("Unable to grab strain data")
```

## Filtering 

```python
trigger_params_file = f"../../benchmarking/non_precessing_random/injections_XPHM_2000_fiducial_pop/injection_{idx}/trigger_parameters.json"
approximant = "IMRPhenomXPHM"
trigger_parameters = io.load_trigger_parameters_from_file(
        trigger_params_file, approximant
    )
```

```python
minimum_data_length = 16
strain, strain_f = io.load_strain_data_from_file(
        trigger_parameters, _strain, _channels, f_low,
        f_high, minimum_data_length=minimum_data_length
    )
```

```python
time_window = 0.1
metric_directions = ['chirp_mass', 'symmetric_mass_ratio', 'chi_align']
peak_finder = 'scipy'
peak_parameters, event_snr = find_peak(
    trigger_parameters, strain_f, psd, approximant,
    f_low, time_window,  dx_directions=metric_directions,
    method=peak_finder
)
```

```python
for param in ["ra", "dec"]:
    if param in trigger_parameters.keys():
        peak_parameters[param] = trigger_parameters[param]
trig_start = trigger_parameters["time"] - time_window
trig_end = trigger_parameters["time"] + time_window

event_snr, _sigma = calculate_snrs_and_sigma(
    peak_parameters, psd, approximant, strain_f, f_low,
    trig_start, trig_end
)
```

```python
peak_parameters.add_fixed("sigma", _sigma)

_snrs, overlaps = calculate_subdominant_snr(
    peak_parameters, psd, approximant, strain_f, f_low,
    trig_start, trig_end
)
event_snr.update(_snrs)
```

```python
if waveforms.precessing_approximant(approximant):
    if not any(peak_parameters.get(p, 0.) for p in ["chi_p", "chi_p2"]):
        _snrs, _overlaps = {"prec": 0.}, {"prec": 0.}
    else:
        _snrs, _overlaps = calculate_precession_snr(
            peak_parameters, psd, approximant, strain_f, f_low,
            trig_start, trig_end
        )
    overlaps.update(_overlaps)
    event_snr.update(_snrs)

# add the overlaps of event_snr
event_snr["overlaps"] = overlaps
# add trigger time
event_snr["trigger_time"] = trigger_parameters["time"]
```

```python
snr_threshold = 4.0
snrs = event_snr
template_parameters = {
    k: peak_parameters[k][0] for k in metric_directions
}
for k in ["ra", "dec"]:
    if k in peak_parameters.keys():
        template_parameters[k] = peak_parameters[k][0]
data_from_matched_filter = {
    "template_parameters": template_parameters,
    "snrs": snrs,
    "sigma": peak_parameters['sigma'][0]
}
pe_result = result.Result(
    f_low=f_low, psd=psd,
    approximant=approximant,
    snr_threshold=snr_threshold,
    data_from_matched_filter=data_from_matched_filter
)
```

```python
# generate the metric
pe_result.generate_metric(metric_directions, dominant_snr=snrs['22'], tolerance=0.01, max_iter=5)
```

```python
# Generate samples distributed according to the metric we just calculated
npts = int(1e6)
_ = pe_result.generate_samples_from_metric(npts=npts)
print(f"{pe_result.samples_dict.number_of_samples} points generated ")
```

```python
pe_result.generate_samples_from_sky(npts=npts)
```

```python
pe_result.reweight_samples(pe.component_mass_prior_weight, 
                          dx_directions=pe_result.metric.dx_directions)
print(f"{pe_result.samples_dict.number_of_samples} samples remaining after mass_ratio prior reweighting")

pe_result.reweight_samples(pe.isotropic_spin_prior_weight,
                          dx_directions=pe_result.metric.dx_directions)
print(f"{pe_result.samples_dict.number_of_samples} samples remaining after chi_p prior reweighting")
```
