```python
filenames = [filename1, filename2, filename3, filename4, filename5]
```

```python
data_snr = pd.DataFrame()
for filename in filenames:
    data = pd.read_csv(
        filename,
        sep=' '
    )
    data_snr = pd.concat([data_snr, data], ignore_index=True)
```

```python
data_snr
```

```python
df_to_plot = data_snr.dropna()
```

```python
df_to_plot
```

```python
df_to_plot.keys()
```

```python
snr_total = np.sqrt(
    df_to_plot['net_snr_1'] ** 2 + df_to_plot['net_snr_-1'] ** 2 + \
    df_to_plot['net_snr_2'] ** 2 + df_to_plot['net_snr_0'] ** 2
)
snr_harms = np.sqrt(
    df_to_plot['net_snr_1'] ** 2 + df_to_plot['net_snr_-1'] ** 2 + \
    df_to_plot['net_snr_2'] ** 2
)
```

```python
total_mass = df_to_plot['mass_1']+df_to_plot['mass_2']
mass_ratio = df_to_plot['mass_1']/df_to_plot['mass_2']
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(
    df_to_plot['eccentricity_10Hz'], (snr_harms/df_to_plot['net_snr_0']), 
    c=total_mass, s=50 
)
plt.xlabel(r'$e_{\rm{10}}$')
plt.ylabel(r'$\rho_{\rm{perp}}/\rho_0$')
plt.colorbar(label=r'$M_{\rm{total}}$')
# plt.yscale('log')
# plt.xscale('log')
plt.ylim(0,1)
plt.show()
```

```python
idx = snr_harms >= 3
idx.sum()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(
    df_to_plot['eccentricity_10Hz'][idx], (snr_harms/df_to_plot['net_snr_0'])[idx], 
    c=total_mass[idx], s=50 
)
plt.xlabel(r'$e_{\rm{10}}$')
plt.ylabel(r'$\rho_{\rm{perp}}/\rho_0$')
plt.colorbar(label=r'$M_{\rm{total}}$')
plt.yscale('log')
plt.xscale('log')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(
    df_to_plot['eccentricity_10Hz'][idx], df_to_plot['net_snr_0'][idx], 
    s=50, label='0'
)
plt.scatter(
    df_to_plot['eccentricity_10Hz'][idx], snr_harms[idx], 
    s=50, label='(1, -1, 2)'
)
plt.xlabel(r'$e_{\rm{10}}$')
plt.legend()
plt.ylabel(r'$\rho$')
plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
# plt.scatter(
#     df_to_plot['eccentricity_10Hz'][idx], df_to_plot['net_snr_0'][idx], 
#     s=50, label='0'
# )
plt.scatter(
    df_to_plot['eccentricity_10Hz'][idx], mass_ratio[idx], c=snr_harms[idx], 
    s=total_mass[idx] * 1.5
)
plt.xlabel(r'$e_{\rm{10}}$')
plt.ylabel(r'$q$')
# plt.legend()
plt.colorbar(label=r'$\rho_{\rm{perp}}$')
# plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
for mode in [0, 1, -1, 2]:
    plt.scatter(
        df_to_plot['eccentricity_10Hz'][idx], df_to_plot[f'net_snr_{mode}'][idx], 
        s=50, label=mode
    )
plt.xlabel(r'$e_{\rm{10}}$')
plt.legend()
plt.ylabel(r'$\rho$')
plt.yscale('log')
plt.show()
```

```python
# Set up bins and common kwargs
bins = np.linspace(-3.5, np.log10(df_to_plot['eccentricity_10Hz'].max()), 30)
kwargs = dict(bins=bins, histtype='step', density=True, lw=2)

hist_all = plt.hist(
    np.log10(data_snr['eccentricity_10Hz']), label=r'CMC', 
    **kwargs
)
hist_all = plt.hist(
    np.log10(df_to_plot['eccentricity_10Hz']), label=r'$\rho_{\rm{Total}} \geq 8$', 
    **kwargs
)
hist_det = plt.hist(
    np.log10(df_to_plot['eccentricity_10Hz'][idx]), label=r'$\rho_{\rm{Perp}} \geq 3$',
    **kwargs
)

plt.xlabel(r'$\log_{10}(e_{\mathrm{10Hz}})$', fontsize=18)
plt.ylabel('Normalized Count', fontsize=18)
plt.title('Eccentricity Distribution', fontsize=20)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
